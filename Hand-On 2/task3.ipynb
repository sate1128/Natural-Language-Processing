```python
import json
import math

data = {} 
with open('movie_title_result.txt') as data_file:  
    data = json.load(data_file)
data
```

```python
pattern = {}
title_pattern = []
title_pattern.append([])
title_pattern.append([])
for line in data:
    for token in line['tokens']:
        single_pattern = ''
        title = ''
        for a in token:
            title += a['word']
            simple = ''
            if a['pos'] == 'NN' or a['pos'] == 'NR' or a['pos'] == 'NT':
                simple = 'N'
            elif a['pos'] == 'VE' or a['pos'] == 'VC' or a['pos'] == 'VV':
                simple = 'V'
            else:
                simple = a['pos']
            single_pattern += simple
            single_pattern += ' '
        if single_pattern in pattern:
            pattern[single_pattern] += 1
        else:
            pattern[single_pattern] = 1
        title_pattern[0].append(title)
        title_pattern[1].append(single_pattern)
        #print(single_pattern)
    #print('-------')
pattern
```

```python
for index,value in enumerate(title_pattern[1]):
    if value == 'N N ':
        print(title_pattern[0][index])
```
