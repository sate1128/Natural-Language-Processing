```python
import json
import math

data = {} 
with open('srt_result.txt') as data_file:  
    data = json.load(data_file)
data
```

```python
word1='喜欢'
word2='妳'

pos1=''
pos2=''

num1 = 0 
num2 = 0
num = 1349
nn = 0
```

```python
import math
for line in data:
    flag1=0
    flag2=0
    for token in line['tokens']:
        for a in token:
            if a['word']==word1 and a['pos']=='VV':
                flag1=1
                pos1 = a['pos']
                num1+=1
            elif a['word']==word2 and a['pos']=='NR':
                flag2=1
                pos2 = a['pos']
                num2+=1
        if flag1==1 and flag2==1:
            nn+=1
print(num,num1,num2,nn)
print(pos1,pos2)
p1=num1/num
p2=num2/num
p12=nn/num
print(p1,p2)
print(math.log2(p12/(p1*p2)))
```
