```python
from pycorenlp import StanfordCoreNLP
import json
from hanziconv import HanziConv
#fantasy horror romance sci_fi war
nlp = StanfordCoreNLP('http://localhost:9000')
f = open('title_war.txt', 'r', encoding='utf-8')
content = f.read()
f.close()
content = content.split('\n')
s = []
for i in content:
    s.append(HanziConv.toSimplified(i))


whole_data = []  
for i,sentence in enumerate(s):
    single_data = {}
    single_data['column'] = []
    single_data['tokens'] = []
    single_data['parse'] = []
    single_data['basicDependencies'] = []
    single_data['column'] = s[i]
    text = s[i]
    output = nlp.annotate(text, properties={
        'annotators': 'tokenize,ssplit,pos,depparse,parse',
        'outputFormat': 'json'
    })
    for each in output['sentences']:
        single_data['tokens'].append(each['tokens'])
    #print(single_data)
    whole_data.append(single_data)
```

```python
pattern = {}
title_pattern = []
for line in whole_data:
    for token in line['tokens']:
        single_pattern = ''
        title = ''
        tmp = []
        for a in token:
            title += a['word']
            simple = ''
            if a['pos'] == 'NN' or a['pos'] == 'NR' or a['pos'] == 'NT':
                simple = 'N'
            elif a['pos'] == 'VE' or a['pos'] == 'VC' or a['pos'] == 'VV':
                simple = 'V'
            else:
                simple = a['pos']
            single_pattern += simple
            single_pattern += ' '
        if single_pattern in pattern:
            pattern[single_pattern] += 1
        else:
            pattern[single_pattern] = 1
        tmp.append(title)
        tmp.append(single_pattern)
        title_pattern.append(tmp)
        #print(single_pattern)
    #print('-------')
pattern
```

```python
out = ''
with open('parse_war.txt', 'w') as outfile:
    json.dump(out, outfile)
```
