```python
tf=[]
for v in range(300):#300
    count_sum = 0
    now = l3[v]
    #print(now)
    for i in data:#304712
        count = 0
        for j in range(len(data[str(i)]['tokens'])):#sentences in tokens 
            for k in range(len(data[str(i)]['tokens'][j])):#words in sentence
                if (now == data[str(i)]['tokens'][j][k]['word'].lower()):
                    count += 1
                    break;
        count_sum+=count
    #print(count_sum)
    tf.append(math.log10(N/count_sum))
tf
```
