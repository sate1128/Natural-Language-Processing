```python
import json
import math

with open('task4_parse.txt') as data_file:  
     file = json.load(data_file)
file
```

```python
for line in file:
    output=''
    for i,sentence in enumerate(line['basicDependencies']):
        root_value=''
        for word in sentence:
            if(word['governorGloss']=='ROOT'):
                root_value = word['dependent']
        #print(root_value)
        sen_num = []
        for k,word in enumerate(sentence):
            if((word['governor']==root_value)|(word['governorGloss']=='ROOT')):
                sen_num.append(word['dependent']-1)
        sen_num.sort()
        for w in sen_num:   
            output+=line['tokens'][i][w]['word']
            output+=' '
    print(output)
    print('-----')
```
